```python
from urllib.request import urlopen
```

```python
reviews_url = "https://www.imdb.com/title/tt0111161/reviews/?ref_=tt_ql_2"
```

```python
# get request to fetch the data
review_data = urlopen(reviews_url)
print(type(review_data))
```

```python
# read from the webpage
review_html = review_data.read()
print(review_html)
review_data.close()
```

## 2. Parsing data

```python
from bs4 import BeautifulSoup as soup
```

```python
# parse the type of data is to be scrapped
review_soup = soup(review_html, 'html.parser')
print(review_soup)
```

```python
#finding relevent info by html tags
div = review_soup.findAll('div',{'class':"text show-more__control"})
```

```python
div[0]
```

```python
with open('review.txt', 'w', encoding='utf-8') as file: 
    for d in div:
        text = d.text
        text +='\n'
        file.write(text)
        
```
