# Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2 
y2 = x*2 +3
```

```python
plt.style.available
```

```python
plt.style.use("seaborn")
```

```python
# line plot

plt.plot(x, y1, label = "Stock Price1", marker = 'o')
plt.plot(x, y2, label = "Stock Price2", linestyle = 'dashed', marker = '*')
plt.legend()
plt.xlabel("Time Stamp")
plt.ylabel("Price")
plt.title("My line Plot")
plt.show()
```

```python
prices = np.array([1,2,3,4])**3
print(prices)
```

```python
# will be plotted on y-axis
plt.plot(prices)
plt.show()
```

## Scatter Plots

```python
# adjusting the size of the plot
plt.figure(figsize = (6,6))
plt.scatter(x, y1, label = "Stock Price1" , color = 'green' , marker = '^')
plt.scatter(x, y2, label = "Stock Price2", color = 'red', marker = '*')
plt.legend()
plt.xlabel("Time Stamp")
plt.ylabel("Price")
plt.title("My scatter Plot")
plt.show()
```

# Bar Graphs

```python
#plt.style.use('dark_background')
plt.style.use('seaborn')
x_coordinates = np.array([0,1,2]) * 2
plt.bar(x_coordinates - 0.25,[10,20,15] , width = 0.5 , label = 'current year' , tick_label = 'Gold Platinum Silver'.split(' '))  # current year
plt.bar(x_coordinates + 0.25,[20,10,12], width = 0.5 , label = 'next year')  #next year
plt.title('Metal Price Comparison')
plt.xlabel('Metal')
plt.ylabel('Price')
plt.xlim(-1,5)
plt.ylim(0,30)  # set y-axis acc to the given range
plt.legend()
plt.show()
```

# Pie Chart

```python
subjects = 'Maths Chem Physics English'.split(' ')
weightage = [20,10,15,5]
plt.pie(weightage,labels = subjects, explode = (0, 0, 0.1, 0) , autopct = '%1.1f%%' , shadow = 'True')
plt.show()
```

# Histogram

```python
# return values with standard normal distribution ( mean = 0 , standard daviation = 1)
xsn = np.random.randn(100)
sigma = 5
u = 70
X =  np.round(xsn*sigma + u)   # gives values having mean = 60 , standard deviation  = 5

X2 = np.round(xsn*5 + 60)
```

```python
print(X, X2)
```

```python
# opacity factor to see where the two distribution overlap
plt.hist(X2  , alpha = 0.8, label = 'Maths')
plt.hist(X , alpha = 0.5, label = 'Physics')
plt.xlabel("Marks range")
plt.ylabel('frequency count of students')
plt.show()
```

# Data visualization challange
Given a movie dataset , task is to make a visualisation that -
- Plot the length of the Movie Title Names on x-axis
- frequency counts on y-axis the number of movies having 'x' characters in their title

```python
import pandas as pd
df = pd.read_csv('files/movie_metadata.csv')
df
```

```python
df.columns
```

```python
movie_title = list(df['movie_title'])
type(movie_title)
```

```python
movie_title[:5]
```

```python
freq_title = {}

for title in movie_title:
    length = len(title)
    
    if freq_title.get(length) is None:
        freq_title[length] = 1
    else:
        freq_title[length] += 1
```

```python
freq_title
```

```python
x = np.array(list(freq_title.keys()))
```

```python
print(x)
```

```python
y = np.array(list(freq_title.values()))
```

```python
print(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("length of movie title")
plt.ylabel("freq count")
plt.title("movie data visualization")
plt.show()
```

# SEABORN Plots

```python
import seaborn as sns
import numpy as np
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data
```

# Bar Plot 
- bivariate

```python
tips_data[tips_data['sex'] == 'Female']['tip'].mean()
```

```python
tips_data[tips_data['sex'] == 'Male']['tip'].mean()
```

```python
# gives bar plot segregate sex column and represent their mean value as height
sns.barplot( x = 'sex', y = 'tip', data = tips_data)
```

```python
sns.barplot( x = 'sex', y = 'tip', data = tips_data , hue = 'time' )
```

```python
# plot standard deviation
sns.barplot( x = 'sex', y = 'tip', data = tips_data , estimator= np.std )
```

# Count Plot
- bar plot that doesn't plot any statistical value but the count of different values 
- univariate

```python
sns.countplot(x = 'sex', data = tips_data)
```

```python
sns.countplot(x = 'day', data = tips_data , hue = 'sex')
```

```python
sns.countplot(x = 'day', data = tips_data)
```

# Box Plot
- quartiles - 25 , 50 , 75 
- outliers
- bivariate


```python
sns.boxplot( x = 'sex', y = 'tip', data = tips_data)
```

# Violin Plot
- left skewed , right skewed properties 
- bivariate plot

```python
# kernal density estimation 
sns.violinplot( x = 'day' , y = 'tip',  hue = 'sex', data = tips_data)
```

# Distribution Plot
- univariate plot

```python
# can specify bins .. this is kinda histogram   can be kde = False 
sns.distplot(tips_data['tip'])
```

# KDE plot
- to get where the data is densly distributed
- univariate plot

```python
sns.kdeplot(tips_data['tip'])
```

# Joint Plot
- bivariate plot
- kinda scatter plot but with more features 

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_data)
```

```python
# dark color represent more data values their
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_data , kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_data , kind = 'reg')
```

# Pair Plot

```python
sns.pairplot(tips_data)
```

```python
sns.pairplot(tips_data , hue = 'sex')
```
