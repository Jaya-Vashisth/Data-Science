```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_test.shape
```

```python
plt.imshow(X_test[90], cmap = 'gray')
```

```python
X = X_test.reshape(-1,28*28)
y = y_test
```

```python
X.shape
```

```python
# Preprocessing
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
sc =  StandardScaler()
```

```python
x_ = sc.fit_transform(X)
```

```python
x_.shape
```

```python
plt.imshow(x_[90].reshape(28,28), cmap = 'gray')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
z_pca = pca.fit_transform(x_)
```

```python
z_pca.shape
```

```python
pca.explained_variance_
```

 Custome PCA

```python
import numpy as np
```

```python
# step -2: compute covariance matrix
```

```python
covar = np.dot(x_.T, x_)
```

```python
covar.shape
```

```python
#step 3 : compute eigen vecctors
```

```python
from numpy.linalg import svd
```

```python
U , S, V = svd(covar)
```

```python
U.shape
```

```python
Ured = U[: , :2]
```

```python
# step 4 : projection of data on new axis
```

```python
z = np.dot(x_, Ured)
```

```python
z.shape
```

Visualize dataset

```python
import pandas as pd
```

```python
new_dataset = np.hstack((z, y.reshape(-1,1)))
```

```python
dataframe = pd.DataFrame(new_dataset, columns = ["PC1", "PC2","label" ])
```

```python
dataframe
```

```python
import seaborn as sns
plt.figure(figsize = (10,10))
fg = sns.FacetGrid(dataframe , hue = 'label', height = 10)
fg.map(plt.scatter, "PC1", "PC2")
plt.legend()
plt.show()
```
