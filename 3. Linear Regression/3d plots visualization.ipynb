# Surface Plots | Data Visualisation
Used to 
 - visualize loss function in machine learning & Deep Learning
 - visualize state or state value Fuction in Reinforcement Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.plot_surface(a, b, a**2+b**2 , cmap = 'rainbow')
plt.show()
```

```python
a = np.arange(-1,1,0.02)
b = a

a, b = np.meshgrid(a,b)
```

```python
fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.plot_surface(a, b, a**2+b**2 , cmap = 'rainbow')
plt.show()
```

```python
# contour plot
fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.contour(a,b,a**2 + b**2 , cmap = 'rainbow')
plt.title('contourplot')
plt.show()
```
