```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
lr = 0.1
x = 0
error = []

plt.plot(X,Y)

for i in range(50):
    
    grad = 2*(x-5)
    x = x - grad*lr
    y = (x - 5)**2
    error.append(y)
    plt.scatter(x,y)
    
```

```python
plt.plot(error)
```
