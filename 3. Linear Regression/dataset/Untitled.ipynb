```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Train.csv')
data
```

```python
data.describe()
```

```python
data_corr = data.corr()
sns.heatmap(data_corr , annot = True)
```

```python
X = data.drop(columns = 'target').values
Y = data['target']
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection = '3d')

#creating plot
ax.scatter3D(X[:,0], X[:,4], color = 'green')
plt.title("3D scatter plot")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
test = pd.read_csv("Test.csv").values
```

```python
y_ = model.predict(X)
model.score(X,Y)
```

```python
ypredicted = model.predict(test)
```

```python
ypredicted 
```

```python
dt = pd.DataFrame(ypredicted, columns=['target' ] )
dt
```
