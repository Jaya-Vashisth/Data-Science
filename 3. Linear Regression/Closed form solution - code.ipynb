# CLosed form solution for linear regression

theta = inverse((X.T)*X) X.T * Y

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
import pandas as pd
```

```python
X, Y = make_regression(n_samples= 400, n_features=1, n_informative=1, noise = 1.8, random_state=11)

print(X.shape, Y.shape)
```

```python
# changing y into matrix
Y = Y.reshape((-1,1))
```

```python
# normalize
X = (X - X.mean())/X.std()

# Visualize
plt.figure()
plt.scatter(X, Y)
plt.title("Normalized Data")
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))
X_ = np.hstack((X, ones))
print(X_.shape)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X, Y):
    
    Y = np.mat(Y)
    firstPart = np.dot(X.T, X)
    secondPart = np.dot(X.T, Y)
    
    theta = np.linalg.pinv(firstPart)*secondPart
                           
    return theta
```

```python
theta = getThetaClosedForm(X_, Y)
print(theta)
```

```python
plt.style.use('seaborn')
plt.figure()
plt.scatter(X, Y)
plt.plot(X, predict(X_, theta), color = 'red', label = 'Predictions')
plt.legend()
plt.title("Normalized Data")
plt.show()
```
